```python
# Download data from 
# https://github.com/aaspip/data/blob/main/viking-ref-000.bin
# 
# This data is from 
# Viking Graben data (many papers)
# 

from pyseistr import binread

data=binread('viking-ref-000.bin',n1=1500,n2=1012) #Data containing multiples

```

```python
from pyseistr import plot2d
import matplotlib.pyplot as plt
plot2d(data,dx=1,dz=0.004,showf=False, close=False)

plt.gca().set_ylabel('Time (s)')
plt.gca().set_xlabel('Shot NO #')
```
