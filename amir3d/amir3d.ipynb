```python
# Download data from 
# https://github.com/aaspip/data/blob/main/AmirSyn3D.mat
# 
# This data is from 
# Nazari Siahsar, M.A., Gholtashi, S., Kahoo, A.R., Chen, W. and Chen, Y., 2017. Data-driven multitask sparse dictionary learning for noise attenuation of 3D seismic data. Geophysics, 82(6), pp.V385-V396.
#

from scipy.io import loadmat

tmp=loadmat('AmirSyn3D.mat')

data=tmp['Data']
datan=tmp['NoisyData']
data_multitask=tmp['DataMultiTask'] #Amir's method
dt=float(tmp['dt'])
```

```python
from pyseistr import plot3d
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))
ax = plt.subplot(projection='3d')
plot3d(data, dx=1,dy=1,dz=dt,vmin=-1,vmax=1, ifnewfig=False, showf=False, close=False)
plt.gca().set_xlabel("X (sample)",fontsize='large', fontweight='normal')
plt.gca().set_ylabel("Y (sample)",fontsize='large', fontweight='normal')
plt.gca().set_zlabel("Time (s)",fontsize='large', fontweight='normal')
plt.title("Clean data",fontsize='large', fontweight='normal')
plt.show()

plt.figure(figsize=(12, 12))
ax = plt.subplot(projection='3d')
plot3d(datan, dx=1,dy=1,dz=dt,vmin=-1,vmax=1, ifnewfig=False, showf=False, close=False)
plt.gca().set_xlabel("X (sample)",fontsize='large', fontweight='normal')
plt.gca().set_ylabel("Y (sample)",fontsize='large', fontweight='normal')
plt.gca().set_zlabel("Time (s)",fontsize='large', fontweight='normal')
plt.title("Noisy data",fontsize='large', fontweight='normal')

plt.figure(figsize=(12, 12))
ax = plt.subplot(projection='3d')
plot3d(data_multitask, dx=1,dy=1,dz=dt,vmin=-1,vmax=1, ifnewfig=False, showf=False, close=False)
plt.gca().set_xlabel("X (sample)",fontsize='large', fontweight='normal')
plt.gca().set_ylabel("Y (sample)",fontsize='large', fontweight='normal')
plt.gca().set_zlabel("Time (s)",fontsize='large', fontweight='normal')
plt.title("Denoised data using Multi-task sparse learning (Amir's method)",fontsize='large', fontweight='normal')

plt.figure(figsize=(12, 12))
ax = plt.subplot(projection='3d')
plot3d(datan-data_multitask, dx=1,dy=1,dz=dt,vmin=-1,vmax=1, ifnewfig=False, showf=False, close=False)
plt.gca().set_xlabel("X (sample)",fontsize='large', fontweight='normal')
plt.gca().set_ylabel("Y (sample)",fontsize='large', fontweight='normal')
plt.gca().set_zlabel("Time (s)",fontsize='large', fontweight='normal')
plt.title("Removed noise using Multi-task sparse learning (Amir's method)",fontsize='large', fontweight='normal')
```
