```python
# Download data from 
# https://github.com/aaspip/data/blob/main/syn_edge2d.bin
# 
# This data is from 
# Fomel, S., 2002. Applications of plane-wave destruction filters. Geophysics, 67(6), pp.1946-1960.
# Gan, S., Wang, S., Chen, Y., Chen, X., Huang, W. and Chen, H., 2016. Compressive sensing for seismic data reconstruction via fast projection onto convex sets based on seislet transform. Journal of Applied Geophysics, 130, pp.194-208.
```

```python
from pyseistr import sigmoid,plot2d,smooth

data=sigmoid(reflectivity=True);

plot2d(data,dx=1,dz=1)

data=smooth(data,rect=[3,1,1])

plot2d(data,dx=1,dz=1)
```
