```python
# Download data from 
# https://github.com/aaspip/data/blob/main/syn_coh2d.mat
# 
# This data is from 
# Zhou, Y., Yang, J., Wang, H., Huang, G. and Chen, Y., 2020. Statistics-guided dictionary learning for automatic coherent noise suppression. IEEE Transactions on Geoscience and Remote Sensing, 60, pp.1-17.
# 
# This is a benchmark data for testing the coherent (or linear) noise attenuation 

from scipy.io import loadmat

data=loadmat("syn_coh2d.mat")
```

```python
from pyseistr import plot2d
d=data['d']
dl=data['dl']
dh=data['dh']

plot2d(dh,dx=1,dz=1)
plot2d(dl,dx=1,dz=1)
plot2d(d,dx=1,dz=1)
```
