```python
# Download data from 
# https://github.com/aaspip/data/blob/main/hyper4_126_32_32.bin
# 
# This data is from 
# Bai, M., Huang, G., Wang, H. and Chen, Y., 2020. Seismic signal enhancement based on the low‚Äêrank methods. Geophysical Prospecting, 68(9), pp.2783-2807.
# 

from pyseistr import binread

hyper=binread("hyper4_126_32_32.bin",n1=126,n2=32,n3=32)
```

```python
from pyseistr import plot3d

plot3d(hyper,dx=1,dy=1,dz=1)
```

```python
import numpy as np
[n1,n2,n3]=hyper.shape
seed=202526
var=0.1
np.random.seed(seed)
var=var*np.abs(hyper).max()
n=var*np.random.randn(n1,n2,n3); #np.random.randn()'s variance is around 1, mean is 0
hypern=hyper+n;

plot3d(hypern, dx=1, dy=1, dz=1)
```

```python
from pydrr import drr3d

d1=drr3d(hypern,0,120,0.004,6,4);

plot3d(d1, dx=1, dy=1, dz=1)
```

```python
from pyseistr import snr

print('SNR of noisy data:', snr(hyper,hypern, 2))
print('SNR of DRR data (no window):', snr(hyper,d1, 2))
```
