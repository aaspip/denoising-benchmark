```python
# Download data from 
# https://github.com/aaspip/data/blob/main/seans_500_180_20.bin
# 
# This data is from, e.g.,
# Gan, S., Wang, S., Chen, Y., Chen, X., Huang, W. and Chen, H., 2016. Compressive sensing for seismic data reconstruction via fast projection onto convex sets based on seislet transform. Journal of Applied Geophysics, 130, pp.194-208.
# ... Any Many Others
# 

from pyseistr import binread

data=binread('seans_500_180_20.bin',n1=500,n2=180,n3=20)
dt=0.004;

data.max()
```

```python
from pyseistr import plot3d
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))
ax = plt.subplot(projection='3d')
plot3d(data, dx=1,dy=1,dz=dt,vmin=-1000,vmax=1000, ifnewfig=False, showf=False, close=False)
plt.gca().set_xlabel("X (sample)",fontsize='large', fontweight='normal')
plt.gca().set_ylabel("Y (sample)",fontsize='large', fontweight='normal')
plt.gca().set_zlabel("Time (s)",fontsize='large', fontweight='normal')
plt.title("Clean data",fontsize='large', fontweight='normal')
plt.show()

```
